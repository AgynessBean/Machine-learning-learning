This small project is aimed at practicing:

--Data analysis and visualization
--Machine learning algorithms: k-means clustering, k-nearest neighbours
--Choosing a Machine Leanring algorithm to predict readmissions in the hospital
--investigating the chosen ML model in terms of most important features selected and the effect of these features in predicting the outcome.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
from pdpbox import pdp, get_dataset, info_plots
import shap
```

```python
data = pd.read_csv('hospital-readmissions-orig.csv')
```

```python
data.head()
```

```python
y = (data['Man of the Match'] == "Yes")
```

```python
feature_names = [i for i in data.columns if data[i].dtype in [np.int64]] #under object feature_names, only include variables that are integers
data.info()
```

```python
feature_names
```

```python
X = data[feature_names]
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
my_model = RandomForestClassifier(random_state=0).fit(train_X,train_y)
```

```python
import eli5
```

```python
from eli5.sklearn import PermutationImportance
```

```python
perm = PermutationImportance(my_model, random_state=1).fit(val_X, val_y)
```

```python
eli5.show_weights(perm, feature_names = val_X.columns.tolist())
```

```python
#try tree model to observe partial dependence
tree_model = DecisionTreeClassifier(random_state=0, max_depth=5, min_samples_split=5).fit(train_X, train_y)
```

```python
from sklearn import tree
import graphviz
```

```python
tree_graph = tree.export_graphviz(tree_model, out_file=None, feature_names=feature_names)
graphviz.Source(tree_graph)
```

```python
# Create the data that we will plot
pdp_goals = pdp.pdp_isolate(model=tree_model, dataset=val_X, model_features=feature_names, feature='Goal Scored')
```

```python
#plot it
pdp.pdp_plot(pdp_goals, 'Goal Scored')
plt.show()
```

```python
feature_to_plot = 'Distance Covered (Kms)'
pdp_dist = pdp.pdp_isolate(model=tree_model, dataset=val_X, model_features=feature_names, feature=feature_to_plot)

pdp.pdp_plot(pdp_dist, feature_to_plot)
plt.show()
```

```python
# Build Random Forest model
rf_model = RandomForestClassifier(random_state=0).fit(train_X, train_y)

pdp_dist = pdp.pdp_isolate(model=rf_model, dataset=val_X, model_features=feature_names, feature=feature_to_plot)

pdp.pdp_plot(pdp_dist, feature_to_plot)
plt.show()
```

```python
# Similar to previous PDP plot except we use pdp_interact instead of pdp_isolate and pdp_interact_plot instead of pdp_isolate_plot
features_to_plot = ['Goal Scored', 'Distance Covered (Kms)']
inter1  =  pdp.pdp_interact(model=tree_model, dataset=val_X, model_features=feature_names, features=features_to_plot)

pdp.pdp_interact_plot(pdp_interact_out=inter1, feature_names=features_to_plot, plot_type='contour', plot_pdp=True)
plt.show()
```

```python
row_to_show = 5
data_for_prediction = val_X.iloc[row_to_show]  # use 1 row of data here. Could use multiple rows if desired
data_for_prediction_array = data_for_prediction.values.reshape(1, -1)


my_model.predict_proba(data_for_prediction_array)
```

```python
import shap  # package used to calculate Shap values

# Create object that can calculate shap values
explainer = shap.TreeExplainer(my_model)

# Calculate Shap values
shap_values = explainer.shap_values(data_for_prediction)
```

```python
shap.initjs()
shap.force_plot(explainer.expected_value[1], shap_values[1], data_for_prediction)
```
